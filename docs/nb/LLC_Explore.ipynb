# Grab one time instance

```python
SST_0 = ds_sst.Theta.isel(time=0, k=0)#, i=slice(1000,2000), j=slice(1000,2000))
SST_0
```

```python
SST_values = SST_0.values
```

```python
list(SST_0.coords.keys())
```

```python
val = SST_values
min1, max1 = np.min(val[np.invert(np.isnan(val))]), np.max(val[np.invert(np.isnan(val))])
min1, max1
```

## Strip coords

```python
drop_coord = []
for key in SST_0.coords.keys():
    if key in ['i','j']:
        continue
    drop_coord.append(key)
drop_coord
```

```python
strip_0 = SST_0.drop_vars(drop_coord)
strip_0
```

```python
xr_0 = strip_0.to_dataset()
xr_0
```

```python
list(xr_0.data_vars.keys())
```

# Write to disk

```python
encoding = {'Theta': {'dtype': 'int16', 'scale_factor': 1e-3, 'add_offset': 10., 'zlib': True, 'missing_value': -32768}}
```

```python
xr_0.to_netcdf('tst.nc', encoding=encoding)
```
