```python
# imports
import numpy as np

import xarray as xr
#import h5py 
#import iris

import xmitgcm.llcreader as llcreader
import xmitgcm
import fsspec

import datetime
```

# Initialize

```python
model = llcreader.ECCOPortalLLC4320Model()
```

# Example

```python
ds = model.get_dataset(varnames=['Theta','U','V','Salt'], k_levels=[0], type='latlon', iter_step=1728)#, iter_step=960)  # Every 12 hours
ds
```

```python
ds.nbytes / 1e12
```

# Grab one time instance

## Grab all three?

```python
ds_0 = ds.isel(time=0, k=0)
```

```python
ds_0
```

```python
# Strip coords
drop_coord = []
for key in ds_0.coords.keys():
    if key in ['i', 'j', 'i_g', 'j_g']:
        continue
    drop_coord.append(key)
ds_0 = ds_0.drop_vars(drop_coord)
```

```python
encoding = {}
encoding['Theta'] = {'dtype': 'int16', 'scale_factor': 1e-3,
                        'add_offset': 10., 'zlib': True, 
                        'missing_value': -32768}
encoding['U'] = {'dtype': 'int16', 'scale_factor': 1e-3,
                        'add_offset': 10., 'zlib': True, 
                        'missing_value': -32768}
encoding['V'] = {'dtype': 'int16', 'scale_factor': 1e-3,
                        'add_offset': 10., 'zlib': True, 
                        'missing_value': -32768}
```

### Write me

```python
ds_0.to_netcdf('tmp.nc', encoding=encoding)
```
