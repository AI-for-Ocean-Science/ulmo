# LLC Coordinates

```python
# imports
import numpy as np

import xarray as xr
#import h5py 
#import iris

import xmitgcm.llcreader as llcreader
import xmitgcm
import fsspec

import datetime
```

# Model object

```python
model = llcreader.ECCOPortalLLC4320Model()
model
```

# Load dataset

```python
ds_sst = model.get_dataset(varnames=['Theta'], k_levels=[0], type='latlon', iter_step=1728)#, iter_step=960)  # Every 12 hours
ds_sst
```

# Coords

```python
ds_sst.XC.variable
```

```python
ds_sst.YC.variable
```

## Values

```python
XC = ds_sst.XC.values
```

```python
XC.min(), XC.max()
```

```python
YC = ds_sst.YC.values
```

```python
YC.min(), YC.max()
```

# Write to disk

## Generate

```python
dumb = np.zeros_like(XC, dtype=int)
```

```python
ds = xr.Dataset(data_vars=dict(zero=(["x", "y"], dumb)), 
                               coords=dict(
                                   lon=(["x", "y"], XC),
                                   lat=(["x", "y"], YC)
                               ))
```

```python
ds
```

## Write

```python
ds.to_netcdf('LLC_coords.nc')
```
