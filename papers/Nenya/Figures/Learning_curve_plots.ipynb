```python
import h5py
```

```python
import matplotlib.pyplot as plt
```

```python
losses_hist_train_file = './LLC_MODIS_2012_curl_losses_train_30_epochs.h5'
losses_hist_valid_file = './LLC_MODIS_2012_curl_losses_valid_30_epochs.h5'
```

```python
with h5py.File(losses_hist_train_file) as f:
    print(f.keys())
    loss_avg_train = f['loss_avg_train'][:]
    loss_step_train = f['loss_step_train'][:]
    loss_train = f['loss_train'][:]
```

```python
with h5py.File(losses_hist_valid_file) as f:
    print(f.keys())
    loss_avg_valid = f['loss_avg_valid'][:]
    loss_step_valid = f['loss_step_valid'][:]
    loss_valid = f['loss_valid'][:]
```

```python
plt.figure()
plt.plot(loss_avg_train, label='train', linewidth=2.5, color='blue')
plt.xlabel('Steps', fontsize=15)
plt.ylabel('Loss', fontsize=15)
plt.title('Learning Curve (step) on training set', fontsize=15)
plt.legend(fontsize=15)
plt.savefig('learning_curve_step_train_30_epochs.png', dpi=100)
```

```python
plt.figure()
plt.plot(loss_avg_valid, label='valid', linewidth=2.5, color='red')
plt.xlabel('Steps', fontsize=15)
plt.ylabel('Loss', fontsize=15)
plt.title('Learning Curve (steps) on validation set', fontsize=15)
plt.legend(fontsize=15)
plt.savefig('learning_curve_step_valid_30_epochs.png', dpi=100)
```

## Epoch Learning curve

```python
plt.figure()
plt.plot(loss_train, label='train_loss', linewidth=2.5, color='blue')
plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Loss', fontsize=15)
plt.title('Learning Curve on training set', fontsize=15)
plt.legend(fontsize=15)
plt.savefig('learning_curve_train_30_epochs.png', dpi=100)
```

```python
plt.figure()
plt.plot(loss_valid[:8], label='valid_loss', linewidth=2.5, color='red')
plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Loss', fontsize=15)
plt.title('Learning Curve on validation set', fontsize=15)
plt.legend(fontsize=15)
plt.savefig('learning_curve_valid_part_30_epochs.png', dpi=100)
```
