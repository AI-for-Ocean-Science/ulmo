# U0 vs. DT

```python
# imports
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

from ulmo import io as ulmo_io
from ulmo.plotting import plotting
```

```python
sns.set_theme()
sns.set_style('whitegrid')
sns.set_context('notebook')
```

# Load up

```python
tbl_file = 's3://modis-l2/Tables/MODIS_L2_std.parquet'
modis_tbl = ulmo_io.load_main_table(tbl_file)
```

## Remove NaN

```python
good = np.isfinite(modis_tbl.LL)
modis_tbl = modis_tbl[good].copy()
```

## Add $\Delta T$

```python
modis_tbl['DT'] = modis_tbl.T90 - modis_tbl.T10
```

```python
modis_tbl['logDT'] = np.log10(modis_tbl.DT)
```

# displot

```python
sns.displot(data=modis_tbl, x='U0', y='DT')
```

```python
sns.displot(data=modis_tbl, x='U0', y='logDT', height=7.)
```
