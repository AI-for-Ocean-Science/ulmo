# Check Model

```python
# imports
import pandas

from ulmo.utils import models as model_utils
```

```python
def count_parameters(model):
    names, params = [], []
    total_params = 0
    for name, parameter in model.named_parameters():
        if not parameter.requires_grad: continue
        param = parameter.numel()
        # Save
        names.append(name)
        params.append(param)
        total_params+=param
    # Table
    df = pandas.DataFrame(dict(Modules=names, Parameters=params))
    return df, total_params
```

# Load up the model

```python
pae = model_utils.load('standard')
```

## Count em -- trainable parameters

```python
df_a, tot_params_a = count_parameters(pae.autoencoder)
```
