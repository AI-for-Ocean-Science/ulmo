# Bias Trends

```python
# imports
from importlib import reload
import numpy as np
import os

import h5py

import pandas

import seaborn as sns
from matplotlib import pyplot as plt

from ulmo.mae import bias
from ulmo.mae import mae_utils
from ulmo import io as ulmo_io
```

```python
%matplotlib widget
```

# Load

```python
bias_tbl = ulmo_io.load_main_table('llc_with_bias.parquet')
```

```python
bias_tbl.keys()
```

# LL?

```python
plt.clf()
fg = sns.displot(data=bias_tbl, x='LL', y='bias', log_scale=(True,False))
fg.set(ylim=(-0.02, 0.07), xlim=(10,2000))
#
plt.ioff()
plt.show()
```

# Mean Temperature?

```python
plt.clf()
fg = sns.displot(data=bias_tbl, x='mean_temperature', y='bias')#, log_scale=(True,False))
fg.set(ylim=(-0.06, 0.11))#, xlim=(10,2000))
#
plt.show()
```

# Latitude

```python
plt.clf()
fg = sns.displot(data=bias_tbl, x='lat', y='bias')#, log_scale=(True,False))
fg.set(ylim=(-0.06, 0.11))#, xlim=(10,2000))
#
plt.ioff()
plt.show()
```
